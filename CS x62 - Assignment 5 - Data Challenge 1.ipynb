```python
# Ensure scikit-learn is installed so imports from sklearn can be resolved
%pip install -q scikit-learn

import numpy as np
from pathlib import Path
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df_test = pd.read_csv('earnings_test_features.csv')
df_test
```

```python
x = df.drop(columns=['WAGE_YEAR4'])
y = df['WAGE_YEAR4']

# Identify numeric and categorical columns
numeric_cols = x.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = [c for c in x.columns if c not in numeric_cols]

# Preprocessing pipelines for both numeric and categorical data
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_cols),
    ('cat', categorical_transformer, categorical_cols)
])

# Define models and their hyperparameter grids for GridSearchCV
models = {
    'Linear Regression': (Ridge(), {'model__alpha': [0.1, 1.0, 10.0]}),
    'CART': (DecisionTreeRegressor(random_state=42), 
                      {'model__max_depth': [5, 10, None],
                       'model__min_samples_leaf': [1, 5]}),
    'KNN': (KNeighborsRegressor(), 
            {'model__n_neighbors': [5, 9, 15],
             'model__weights': ['uniform', 'distance']})
    
}

# Cross-validation and model evaluation setup
cv = KFold(n_splits=5, shuffle=True, random_state=42)
mse_scorer = make_scorer(lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, 
                        y_pred)), greater_is_better=False)
results = []
best_model = None
best_score = float('inf')

# Train and evaluate each model using GridSearchCV
for model_name, (estimator, param_grid) in models.items():
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', estimator)
    ])
    grid_search = GridSearchCV(pipeline, param_grid, scoring=mse_scorer, cv=cv, 
                               n_jobs=-1, refit=True)
    grid_search.fit(x, y)
    mean_mse = -grid_search.best_score_
    results.append((model_name, mean_mse, grid_search.best_params_))
    if mean_mse < best_score:
        best_score = mean_mse
        best_model = grid_search.best_estimator_

print("Model Performance:")
for r in results:
    print(f"Model: {r[0]}, Mean RMSE: {r[1]:.4f}, Best Params: {r[2]}")

# Fit the best model on the entire training data
best_model.fit(x, y)

# Generate predictions on the test set and save to CSV
predictions = best_model.predict(df_test)
pd.DataFrame({'WAGE_YEAR4': predictions}).to_csv('preds.csv', index=False, header=False)
print("Predictions saved to preds.csv")
```

Participants with higher educational levels, especially those with college degrees, had higher predicted incomes, while older individuals earned more than younger participants, due to gained experience. From an ethical perspective, it is crucial to recognize that while these features improve prediction accuracy, they also reflect real-world inequalities. If previous salaries or demographic variables are largely relied on for decision-making, for example, in hiring or loan assessments, they may unintentionally recreate historical inequalities. As a result, these features should only be used for analytical statistics or policy evaluations, and not for individual-level decisions.
